### Replacing Yes/No with 0 and 1

```python
df = df.replace(['No', 'no'], 0)
df = df.replace(['Yes', 'yes'], 1)
#df = df.replace([np.NaN], 'Others')

#df = df.replace(['No internet service'], 0)
#df = df.replace(['No phone service'], 0)
```

## Outliers

```python
df2 = df[['Account Length (in months)', 'Local Calls', 'Local Mins', 
          'Intl Calls', 'Intl Mins', 'Extra International Charges', 
          'Customer Service Calls', 'Avg Monthly GB Download',
          'Extra Data Charges', 'Age', 
          'Monthly Charge', 'Total Charges']]
```

```python
n = 1
plt.figure(figsize=(36,30))
for column in df2.describe().columns:
    plt.subplot(4, 4, n)
    n = n+1
    sns.boxplot(df2[column])
    plt.title(str(column))
    plt.tight_layout()
plt.show()
```

**Removing top 2% of Oultiers in the data**

```python
higher_limit = df['Local Calls'].quantile(0.98)
new_df = df[df['Local Calls'] <= higher_limit]
new_df.shape
```

```python
sns.boxplot(new_df['Local Calls'])
```

```python
higher_limit = new_df['Intl Calls'].quantile(0.98)
new_df = new_df[new_df['Intl Calls'] <= higher_limit]
higher_limit = new_df['Intl Mins'].quantile(0.98)
new_df = new_df[new_df['Intl Mins'] <= higher_limit]
higher_limit = new_df['Extra International Charges'].quantile(0.98)
new_df = new_df[new_df['Extra International Charges'] <= higher_limit]
higher_limit = new_df['Avg Monthly GB Download'].quantile(0.98)
new_df = new_df[new_df['Avg Monthly GB Download'] <= higher_limit]
```

```python
new_df.shape
```

```python
df3 = new_df[['Account Length (in months)', 'Local Calls', 'Local Mins', 
          'Intl Calls', 'Intl Mins', 'Extra International Charges', 
          'Customer Service Calls', 'Avg Monthly GB Download',
          'Extra Data Charges', 'Age', 
          'Monthly Charge', 'Total Charges']]
```

```python
n = 1
plt.figure(figsize=(36,30))
for column in df3.describe().columns:
    plt.subplot(4, 4, n)
    n = n+1
    sns.boxplot(df3[column])
    plt.title(str(column))
    plt.tight_layout()
plt.show()
```

### Correlation with Response Variable

```python
new_df.corrwith(new_df['Churn Label']).plot.bar(
        figsize = (20, 10), title = "Correlation with Churn", fontsize = 15,
        rot = 70, grid = True)
plt.show()
```
